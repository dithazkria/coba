```python
import tensorflow as tf

(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

print("Training data shape:", X_train.shape, y_train.shape)
print("Test data shape:", X_test.shape, y_test.shape)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.imshow(X_train[i], cmap='gray')
    plt.title(f"Label: {y_train[i]}")
    plt.axis('off')

plt.show()
```

```python
import matplotlib.pyplot as plt
sample = 1
image = X_train[sample]

fig = plt.figure
plt.imshow(image, cmap= 'gray')
plt.show()
```

```python
# Normalisasi data ke rentang 0-1
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Mengubah gambar 28x28 menjadi vektor 784
X_train_flattened = X_train.reshape(X_train.shape[0], -1)
X_test_flattened = X_test.reshape(X_test.shape[0], -1)
```

```python
num_clusters = 10

kmeans = KMeans(n_clusters=num_clusters, random_state=42)

kmeans.fit(X_train_flattened)

y_kmeans = kmeans.predict(X_test_flattened)
```

```python
fig, axes = plt.subplots(1, num_clusters, figsize=(10, 3))
for i, ax in enumerate(axes):
    ax.imshow(kmeans.cluster_centers_[i].reshape(28, 28), cmap='gray')
    ax.axis('off')

plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Calculate the silhouette score for the K-Means clustering on the test data\n",
score = silhouette_score(X_test_flattened, y_kmeans)

print("Silhouette Score:", score)
```
